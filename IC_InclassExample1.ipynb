```python
import matplotlib.pyplot as plt 
#Import matplotlib for plotting
```

```python
def N_iterative(N_0, p, time): 
#Exponential growth model
#Define a function for exponential growth 
#N_0 = initial duckweed patch size in m^2
#p = daily growth rate (0.4 for 40% per day)
#time = number of days to simulate
    N_values = [N_0]   
    #Start a list to store patch size over time
    #Begin with the observed initial patch size at day 0
    current_N = N_0  
    #Set the current patch size to the starting value
    for _ in range(time):
    #For each day that passes
        current_N = current_N + p * current_N 
        #Update the patch size using the exponential growth formula
        #Nt+1 = Nt + p*Nt = (1+p)*Nt
        current_N = round(current_N, 3)  
        #Round the patch size to 3 decimal places
        N_values.append(current_N) 
        #Record the updated patch size for this day
    return N_values   
    #Return the full list of patch sizes for each day
```

```python
#Run the function with 35 days and p= .4 with an initial population of .01
result = N_iterative(0.01, 0.4, 35) 
#Initial patch size = 0.01 m**2, daily growth rate = 40%,
#project growth for 35 days
print(result) 
#Print the list of duckweed patch sizes day by day
```

```python
#Plot the results with days on the x-axis and population size on the y-axis
days = list(range(36))  
#Days from 0 to 35 
plt.plot(days, result, marker='o') 
#Plot with markers
plt.xlabel('Days')
#Label the x-axis as "Days"
plt.ylabel('Population Size') 
#Label the y-axis to Population size
plt.title('Population Growth Over Time') 
#Add title to the plot
plt.grid() 
#Add grid 
plt.show() 
#Show the plot
plt.yscale('log')
plt.plot(days, result, marker='o')
plt.xlabel('Days')
plt.ylabel('Population Size (log scale)')
plt.title('Population Growth Over Time (Log Scale)')
plt.grid()
plt.show()
```

```python
#Combine these two graphs into one graph with two y-axes
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(days, result, 'red',marker='o')
ax2.plot(days, result, 'black',marker='o')
ax1.set_xlabel('Days')
ax1.set_ylabel('Population Size', color='black')
ax2.set_ylabel('Log Population Size', color='red')
ax2.set_yscale('log')
plt.title('Population Growth Over Time')
plt.grid()
plt.show()
```

```python
#Change N_exmp to N_exp2 and have 3 conditions based on N_0
#If N_0 < 500, p = .4
#If N_0 >= 500 and N_0 < 1000, p = .2
#if N_0 >= 1000, p=0
def N_exp2(N_0, time):
    N_values = [N_0]
    Current_N = N_0
    for _ in range(time):
        if Current_N < 500:
            p = 0.4
        elif 500 <= Current_N < 1000:
            p = 0.2
        else:
            p = 0
        current_N = Current_N + p * Current_N
        current_N = round(current_N, 3)
        N_values.append(current_N)
    return N_values

```

```python
#Run the function with 35 days and p= .4 with an initial population of .01
result = N_exp2(0.01, 35) 
#Initial patch size = 0.01 m**2, daily growth rate = 40%,
#project growth for 35 days
print(result) 
#Print the list of duckweed patch sizes day by day
days = list(range(36))  
#Days from 0 to 35 
plt.plot(days, result, marker='o') 
#Plot with markers
plt.xlabel('Days')
#Label the x-axis as "Days"
plt.ylabel('Population Size') 
#Label the y-axis to Population size
plt.title('Population Growth Over Time') 
#Add title to the plot
plt.grid() 
#Add grid 
plt.show() 
```
