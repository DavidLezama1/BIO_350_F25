```python
#importing numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt


"""

This is the example format I am going to do

def SIR_model(S0,I0,R0,beta,d_rate,time):
    #Import the variables given to the function
    #loop to go through time period 
    return # 3 variables seperated by comma

#Call the SIR_model and give it S0, I0, R0, beta, d_rate, time and save them into 3 variables


"""

def SIR_model(S0, I0, R0, beta, d_rate, time): # making a function
    # import the variables given to the function
    S = [S0] # this is the number of foxes in the susceptible state initially (999)
    I = [I0] # the number of infected fox initially (1) 
    R = [R0] # number of population that initally have died (0) since it just started 
    
    # loop to go through time period 
    for i in range(time): # making a for loop
        S_next = S[-1] - beta * S[-1] * I[-1] # writing out the equation of St+1 = St − βStIt
        I_next = I[-1] + beta * S[-1] * I[-1] - d_rate * I[-1] # writing out the equation of It+1 = It + βStIt − dIt
        R_next = R[-1] + d_rate * I[-1] # writing out the equation of Rt+1 = Rt + dIt
        
        S.append(S_next) # adding all the S_next values every day
        I.append(I_next) # adding all the I_next values every day
        R.append(R_next) # adding all the R_next values every day
    
    return np.array(S), np.array(I), np.array(R)  # return 3 variables separated by comma


# call the SIR_model and give it S0, I0, R0, beta, d_rate, time
S, I, R = SIR_model(S0=999, I0=1, R0=0, beta=0.001, d_rate=0.5, time=40) # Given values all based on textbook + decided to run it for 40 days

# plotting
plt.plot(S, label="Susceptible") # plotting the susceptible (St)
plt.plot(I, label="Infected") # plotting the infected (It)
plt.plot(R, label="Removed/Dead") # plotting the removed (Rt)
plt.xlabel("Days") #labeled the x-axis
plt.ylabel("Number of Foxes") # labeled the y-axis
plt.legend() # added a legend
plt.show() #show the plot




print ("If only this one fox is currently infected, could the disease eventually spread through the population? If so, how many foxes might eventually die of rabies?") # question being asked
print(" ") #Space between question and answer
print("Yes, the rabies will eventually spread through the population and", round(R[-1],0),"foxes will eventually die of rabies by day 40") #Printing the value of removed/dead at day 40
```

Beta variable - What happens when it is changed up or down? What are reasons it may change?

When the beta variable is increased, each infected fox has a higher chance of spreading rabies to susceptible foxes, 
so the disease spreads more quickly and infects more of the population. When beta is decreased, the spread slows down, and
in some cases, the disease may not spread at all (like mine in which I  I decreased the beta variable to .0001) and remain limited to the first infected fox. Beta may change depending on 
how often foxes come into contact with each other, which can be influenced by population density, seasonal behaviors like 
mating or denning, or environmental barriers such as roads or fences.


d variable - What happens when the death rate is changed? Why might the rate change, or why might it be high/low?

When the death rate is higher, infected foxes die more quickly, which reduces the time they have to spread rabies, leading to a smaller epidemic. When the death rate is lower, infected foxes survive longer, which gives them more opportunities to transmit the disease and increases the likelihood that more of the population will eventually be infected. The death rate might change based on biological differences among foxes, such as age or immune system, environmental stressors like food scarcity or co-existing diseases, or differences in the severity of the rabies strain that affect how quickly it kills an infected animal.
