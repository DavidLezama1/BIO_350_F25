```python
#Import "population_Data.csv" into a pandas DataFrame
import pandas as pd
pop_data = pd.read_csv("population_Data.csv")
pop_data.head()
```

```python
#Plot the population daa with "Days" on the x-axis and "Population" on the y-axis
import matplotlib.pyplot as plt
plt.plot(pop_data["Days"], pop_data["Population"], marker='o', color='black', linestyle='-', markerfacecolor='black', markeredgecolor='black')
plt.axhline(y=276, color='green', linestyle='--', label='y=276')
plt.xlabel("Days")
plt.ylabel("Population")
plt.title("Population over Days")
plt.grid(True)
plt.legend()
plt.xlim(0, pop_data["Days"].max())
plt.ylim(0, 350)
plt.show()
pop_data = pd.read_csv("population_Data.csv")
pop_data.head()
```

```python
#average the population from day 17 to day 60
average_population = pop_data[(pop_data["Days"] >= 17) & (pop_data["Days"] <= 60)]["Population"].mean()
average_population
```
