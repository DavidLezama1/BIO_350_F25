```python
#Import "population_Data.csv" into a pandas DataFrame
import pandas as pd #import pandas library
pop_data = pd.read_csv("population_Data.csv") #read the csv file
pop_data.head() #display the first 5 rows of the DataFrame
```

```python


#Plot the population data with "Days" on the x-axis and "Population" on the y-axis
import matplotlib.pyplot as plt #import matplotlib library
plt.plot(pop_data["Days"], pop_data["Population"], marker='o', color='black', linestyle='-', markerfacecolor='black', markeredgecolor='black') #plot the data
plt.axhline(y=276, color='green', linestyle='--', label='y=276') #add horizontal line at y=276
plt.xlabel("Days") #label x-axis
plt.ylabel("Population") #label y-axis
plt.title("Population over Days") #title of the plot
plt.grid(True) #add grid
plt.legend() #add legend
plt.xlim(0, pop_data["Days"].max()) #set x-axis limits
plt.ylim(0, 350) #set y-axis limits
plt.show() #display the plot

```

```python
#Average the population from day 17 to day 60
average_population = pop_data[(pop_data["Days"] >= 17) & (pop_data["Days"] <= 60)]["Population"].mean() #calculate the average population
average_population #display the average population
```
