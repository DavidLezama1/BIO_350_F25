```python
# Create an algorthm that calculates the population of 3 ages of beetle population and the total population over 50 years
# The inital first year population is 500 and the inital second year population is 500
# On average 1 beetle produces 1 offspring in second year and 5 offspring in third year
# The first year beetles survive 30% to second year and 50% of second year bettles survive to third year

import numpy as np
import matplotlib.pyplot as plt

#Create a definition to calculate and return the population of each age group and the total population for each year
def beetle_population(years,inital_first_year,inital_second_year):
    # Set explanations of parameters
    """
    years (int):Number of years to calculate the population for
    inital_first_year (int): Inital population of first year beetles
    inital_second_year (int): Inital population of second year beetles
    Returns:
    first_year_population (list): population of first year beetles for each year
    second_year_population (list): population of second year beetles for each year
    third_year_population (list): population of third year beetles for each year
    total_population (list): total beetle population for each year
    """
    #Complete the algorithm using matrix multiplication
    # Create a matrix to represent the transition of beetle population between age groups
    A = np.array([[0, 1, 5],
                  [0.3, 0, 0],
                  [0, 0.5, 0]], dtype=float)

    inital_population_vector = np.array([inital_first_year, inital_second_year, 0], dtype=float)

    # Initalize the population lists with the initial populations
    first_year_population = []
    second_year_population = []
    third_year_population = []
    total_population = []

    first_year_population.append(inital_population_vector[0])
    second_year_population.append(inital_population_vector[1])
    third_year_population.append(inital_population_vector[2])
    total_population.append(float(inital_population_vector.sum()))

    n_t = inital_population_vector
    for year in range(years):
        n_t = np.dot(A, n_t)
        first_year_population.append(n_t[0])
        second_year_population.append(n_t[1])
        third_year_population.append(n_t[2])
        total_population.append(float(n_t.sum()))

    # Returns the populations of each age group and the total population for each year
    return first_year_population, second_year_population, third_year_population, total_population


# Call the function to calcualte the populations for 150 years with the inital population population is 500 and the inital second year population is 500
first_year_pop, second_year_pop, third_year_pop, total_pop = beetle_population(150, 500, 500)

# Plot the populations of each age group and the total population over time
years_range = range(len(first_year_pop))

plt.figure(figsize=(10, 6))
plt.plot(years_range, first_year_pop, label='First Year')
plt.plot(years_range, second_year_pop, label='Second Year')
plt.plot(years_range, third_year_pop, label='Third Year')
plt.plot(years_range, total_pop, label='Total Population', linestyle='--', color='black')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Beetle Population by Age Group Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create an algorthm that calculates the population of 3 ages of beetle population and the total population over 50 years
# The inital first year population is 500 and the inital second year population is 500
# On average 1 beetle produces 1 offspring in second year and 5 offspring in third year
# The first year beetles survive 30% to second year and 50% of second year bettles survive to third year

import numpy as np
import matplotlib.pyplot as plt

#Create a definition to calculate and return the population of each age group and the total population for each year
def beetle_population(years,inital_first_year,inital_second_year):
    # Set explanations of parameters
    """
    years (int):Number of years to calculate the population for
    inital_first_year (int): Inital population of first year beetles
    inital_second_year (int): Inital population of second year beetles
    Returns:
    first_year_population (list): population of first year beetles for each year
    second_year_population (list): population of second year beetles for each year
    third_year_population (list): population of third year beetles for each year
    total_population (list): total beetle population for each year
    """
    #Complete the algorithm using matrix multiplication
    # Create a matrix to represent the transition of beetle population between age groups
    trap_rate = .05
    A = np.array([[0, 1, 5],
                  [0.3, 0, 0],
                  [0, 0.5, 0]]) * (1 - trap_rate)

    inital_population_vector = np.array([inital_first_year, inital_second_year, 0], dtype=float)

    # Initalize the population lists with the initial populations
    first_year_population = []
    second_year_population = []
    third_year_population = []
    total_population = []

    first_year_population.append(inital_population_vector[0])
    second_year_population.append(inital_population_vector[1])
    third_year_population.append(inital_population_vector[2])
    total_population.append(float(inital_population_vector.sum()))

    n_t = inital_population_vector
    for year in range(years):
        n_t = np.dot(A, n_t)
        first_year_population.append(n_t[0])
        second_year_population.append(n_t[1])
        third_year_population.append(n_t[2])
        total_population.append(float(n_t.sum()))

    # Returns the populations of each age group and the total population for each year
    return first_year_population, second_year_population, third_year_population, total_population


# Call the function to calcualte the populations for 150 years with the inital population population is 500 and the inital second year population is 500
first_year_pop, second_year_pop, third_year_pop, total_pop = beetle_population(150, 500, 500)

# Plot the populations of each age group and the total population over time
years_range = range(len(first_year_pop))

plt.figure(figsize=(10, 6))
plt.plot(years_range, first_year_pop, label='First Year')
plt.plot(years_range, second_year_pop, label='Second Year')
plt.plot(years_range, third_year_pop, label='Third Year')
plt.plot(years_range, total_pop, label='Total Population', linestyle='--', color='black')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Beetle Population by Age Group Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

1. What difference do you see between the two plots (no traps vs with traps)?

The plot without traps shows all age groups of the beetle population increasing at a steady rate, while the plot with traps shows a steady decline in the population over time. The traps clearly prevent the population from recovering, leading to a much smaller number of beetles by the end of the 150 years.

2. How much of an effect did setting up traps have on the beetle population over 150 years?

Setting up traps had a major impact because even catching only five percent of the adult survivors each year greatly reduced the beetle population. Over 150 years, the traps changed the long-term trend from slow growth to continuous decline, eventually bringing the beetle population to extinction.
