 Plot the data showing the frequency of your data with the probability density on the y-axis

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from scipy.stats import norm

style.use('dark_background')  

# The biological scenario
# Assume we measured the shoulder heights (in meters) of 200 adult male giraffes.
# Giraffe height is roughly normally distributed around a mean of 5.2 m with std dev 0.3 m.
np.random.seed(42) # Randow values
heights = np.random.normal(loc=5.2, scale=0.3, size=200)

# We are plotting the histogram 
plt.figure(figsize=(10,6))
count, bins, ignored = plt.hist(
    heights, bins=15, density=True, color='white', alpha=0.6, edgecolor='purple', label='Observed data'
)

# Plotting the Normal curve 
x = np.linspace(min(heights), max(heights), 100)
pdf = norm.pdf(x, loc=5.2, scale=0.3)
plt.plot(x, pdf, 'cyan', linewidth=3, label='Normal distribution fit')

plt.title('Distribution of Adult Male Giraffe Heights', fontsize=18)
plt.xlabel('Shoulder Height (m)', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
